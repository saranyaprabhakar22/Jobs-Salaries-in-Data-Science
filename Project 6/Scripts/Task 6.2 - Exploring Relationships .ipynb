# 3. Creating Correlation matrix heatmap

```python
# Converting employee_residence to numerical

sub['employee_residence'] = len.fit_transform(sub['employee_residence'])
sub['employee_residence'].unique()
```

```python
# Converting experience_level to numerical

sub['experience_level'] = len.fit_transform(sub['experience_level'])
sub['experience_level'].unique()
```

```python
sub.corr()
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Creating the correlation heatmap
corr = sns.heatmap(sub.corr(), annot = True, ax = ax)
```

```python
corr.figure.savefig(os.path.join(path, 'Visualizations', 'corr.png'))
```

# From the above correlation heatmap, variables that are positively correlated with the salary are company_location, employee_residence, experience_level, job_title and salary_currency.

# Creating categorical plot

```python
# Create histogram to visualize the distribution of the variables
sns.histplot(df_jobs['salary_in_usd'], bins = 20, kde = True)
```

```python
# Creating salary category for categorical plot
df_jobs.loc[df_jobs['salary_in_usd'] < 100000, 'salary_category'] = 'Average Salary'
```

```python
df_jobs.loc[(df_jobs['salary_in_usd'] >= 100000) & (df_jobs['salary_in_usd'] < 220000) , 'salary_category'] = 'Medium Salary'
```

```python
df_jobs.loc[df_jobs['salary_in_usd'] >= 220000, 'salary_category'] = 'High Salary'
```

```python
df_jobs['salary_category'].value_counts(dropna = False)
```

```python
# Creating categorical plot with job_title

sns.set(style="ticks")
cat_plot_title = sns.catplot(x='job_title', y='salary_in_usd', hue="salary_category", data=df_jobs)
```

```python
# Creating categorical plot with company_location

sns.set(style="ticks")
cat_plot_loc = sns.catplot(x='company_location', y='salary_in_usd', hue="salary_category", data=df_jobs)
```

```python
# Creating categorical plot with experience_level

sns.set(style="ticks")
cat_plot_exp = sns.catplot(x='experience_level', y='salary_in_usd', hue="salary_category", data=df_jobs)
```

```python
# Creating categorical plot with employment_type

sns.set(style="ticks")
cat_plot_typ = sns.catplot(x='employment_type', y='salary_in_usd', hue="salary_category", data=df_jobs)
```

```python
# Creating categorical plot with work_setting

sns.set(style="ticks")
cat_plot_setting = sns.catplot(x='work_setting', y='salary_in_usd', hue="salary_category", data=df_jobs)
```

```python
# Creating categorical plot with company_size

sns.set(style="ticks")
cat_plot_siz = sns.catplot(x='company_size', y='salary_in_usd', hue="salary_category", data=df_jobs)
```

# From the above plot, the number of professionals working in the medium-sized companies are higher and the interesting part was they are also paid better than the professionals in the large companies.

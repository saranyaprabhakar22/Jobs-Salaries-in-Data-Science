# Checking for mixed data types

```python
for col in df_jobs.columns.tolist():
    weird = (df_jobs[[col]].map(type) != df_jobs[[col]].iloc[0].apply(type)).any(axis =1)
    if len (df_jobs[weird]) > 0:
        print(col)
```
